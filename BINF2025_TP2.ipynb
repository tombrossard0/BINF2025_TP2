Q9.	Donnez l’ensemble des ORFs présents dans le génome de l’AAV2. Présentez vos résultats au format FASTA où pour chaque ORF vous fournissez les informations :
```MARKDOWN
> pos début, pos fin, brin
séquence
```
où "pos début" est la position du premier nucléotide du codon START et "pos fin" est la position du dernier nucléotide du codon STOP (les positions sont comptées depuis le début du brin), brin est l'orientation du brin + ou - (+ est la séquence de base; - est le brin complément inverse) et séquence est simplement la séquence.

````python
import numpy as np

file = open("NC_001401.2.fna")
file.readline()
brin = file.read().replace('\n', '')

def compl(brin):
    res = ""
    for c in brin:
        if c == 'T':
            return 'A'
        if c == 'A':
            return 'T'
        if c == 'G':
            return 'C'
        return 'G'

def isEndCodon(codon):
    return genecode[codon] == '_'

def isStartCodon(codon):
    return genecode[codon] == 'M'

def getSeqs(brin):
    start_positions = []
    
    # Search all start positions
    for i in range(len(brin)):
        if len(brin) - i < 3:
            break
    
        codon = brin[i] + brin[i + 1] + brin[i + 2]
    
        if isStartCodon(codon):
            start_positions.append((i, codon))
        i += 3
    
    valids = []
    
    # For each posisions, check if they have an End
    for (start_pos, start_codon) in start_positions:
        pos = start_pos
        while pos + 2 < len(brin):
            codon = brin[pos] + brin[pos + 1] + brin[pos + 2]
            if isEndCodon(codon):
                valids.append((start_pos, pos))
                break
            pos += 3
    
    sequences = []
    for (start, end) in valids:
        seq = ""
        for i in range(end - start):
            seq += brin[start + i]
        sequences.append(seq)
    return sequences

# For each, check if they have a complement
cadr = getSeqs(brin) + getSeqs(brin[1:]) + getSeqs(brin[2:])
#cadrI = getSeqs(brin, True) + getSeqs(brin[1::], True) + getSeqs(brin[2::], True)

print(len(cadr))
````
